### Question 12

```python
# import train_test_split and linear regression model for splitting the dataset and training our model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X = dataset[['T2']]
y = dataset['T6']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=2)

linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
pred_values = linear_reg.predict(X_test)
```

```python
# R-Squared Score
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, pred_values)
round(r2_score, 2) 
```

### Question 13

```python
df_1 = dataset.drop(columns = ['date', 'lights'], axis=1)

# Normalization
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()

norm_data = pd.DataFrame(Scaler.fit_transform(df_1), columns=df_1.columns)
```

```python
x_1 = norm_data.drop(columns=['Appliances'])
y_1 = norm_data['Appliances']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_1, y_1, test_size=0.3, random_state=42)
```

```python
# Regression
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
predicted_values = linear_model.predict(x_test)
```

```python
# MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

### Question 14

```python
# to get RSS
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

### Question 15

```python
# to get Root Mean Squared Error i.e RMSE
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

### Question 16

```python
# to get the R-Squared or also called Coefficient of Determination
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predicted_values)
round(r2, 2) 
```

### Question 17

```python
# Feature Weights from linear model

def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
    
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
linear_model_weights
```

```python
# to get the the lowest and highest weights respectively
display(linear_model_weights.nsmallest(n=1,columns='Linear_Model_Weight'))
display(linear_model_weights.nlargest(n=1,columns='Linear_Model_Weight'))
```

```python
# LASSO Regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)

Pred_2 = lasso_reg.predict(x_test)

print(np.sum(lasso_reg.coef_ != 0))
```
